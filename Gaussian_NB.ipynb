```python
from sklearn import datasets
iris = datasets.load_iris()
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(iris.data, iris.target).predict(iris.data)
print("Number of mislabeled points out of a total %d points : %d"
             % (iris.data.shape[0],(iris.target != y_pred).sum()))
```

```python
import numpy as np
data = iris.data
target = iris.target
```

```python
def calc_para(data):
    mus = np.zeros((4,3))
    sigmas = np.zeros((4,3))
    for i in range(4):
        for j in range(3):
            # i refer to the index of feature, j refer to the index of class
            xi = data[:,i]
            xij = xi[target==j]
            muij = np.mean(xij)
            sigmaij = np.std(xij)
            mus[i,j] = muij
            sigmas[i,j] = sigmaij 
    total = data.shape[0]
    portion = []
    for j in range(3):
        c = np.sum(target==j)
        portion.append(c)
    portion = np.array(portion)
    return sigmas,mus,portion/total
```

```python
sigmas,mus,portion = calc_para(data)
```

```python
def predict(sample,sigmas,mus,portion):
    scores = []
    for j in range(3):
        score = 1
        for i in range(4):
            prob0 = 1/sigmas[i,j]*np.e**(-(sample[i]-mus[i,j])**2/(2*sigmas[i,j]**2))
            score *= prob0
        score *= portion[j]
        scores.append(score)
    scores = np.array(scores)
    return np.argmax(scores)
```

```python
prediction = [predict(data[i,:], sigmas, mus, portion) for i in range(data.shape[0])]
```

```python
print("Number of mislabeled points out of a total %d points : %d"
             % (iris.data.shape[0],(iris.target != prediction).sum()))
```
